```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite
from networkx.drawing.layout import bipartite_layout
```

```python
ratings = pd.read_csv('100k/u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
movies = pd.read_csv('100k/u.item', sep='|', names=['movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL', 
                                                       'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 
                                                       'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 
                                                       'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'], 
                                                       encoding='latin-1')
users = pd.read_csv('100k/u.user', sep='|', names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])
```

```python
# Calculate the number of reviews per user
user_review_counts = ratings['user_id'].value_counts()

# Display the first few user review counts
print(user_review_counts.head())
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the number of reviews per user
plt.figure(figsize=(10, 6))
user_review_counts.hist(bins=50)
plt.title('Distribution of Number of Reviews per User')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate mean and standard deviation of the number of reviews per user
mean_reviews = user_review_counts.mean()
std_reviews = user_review_counts.std()

print(f"Mean number of reviews: {mean_reviews}")
print(f"Standard deviation of reviews: {std_reviews}")
```

```python
# Define the threshold for outliers (3 standard deviations from the mean)
threshold_low = mean_reviews - 2 * std_reviews
threshold_high = mean_reviews + 2 * std_reviews

# Identify users to keep (within the threshold)
filtered_users = user_review_counts[(user_review_counts >= threshold_low) & (user_review_counts <= threshold_high)].index

# Filter the ratings DataFrame to include only the users within the threshold
filtered_ratings = ratings[ratings['user_id'].isin(filtered_users)]

print(f"Number of users after filtering: {len(filtered_users)}")
```

```python
# Convert timestamp to datetime
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')

user_review_counts = ratings['user_id'].value_counts()
active_users = user_review_counts[user_review_counts >= 3].index

filtered_ratings = ratings[ratings['user_id'].isin(active_users)]
print(f"Number of active users: {len(active_users)}")

from datetime import timedelta

# Define a threshold for the maximum allowable time span (e.g., 1 year)
time_span_threshold = timedelta(days=365)

# Calculate the time span for each user's reviews
def user_time_span(user_id):
    user_ratings = filtered_ratings[filtered_ratings['user_id'] == user_id]
    time_span = user_ratings['timestamp'].max() - user_ratings['timestamp'].min()
    return time_span

# Filter out users with reviews spread over too long of a time period
filtered_users = [user_id for user_id in active_users if user_time_span(user_id) <= time_span_threshold]

final_filtered_ratings = filtered_ratings[filtered_ratings['user_id'].isin(filtered_users)]
print(f"Number of filtered users: {len(filtered_users)}")
```

```python
# Process the data
user_movie_interactions = ratings[['user_id', 'item_id']]

# Create a bipartite graph
B = nx.Graph()
users = user_movie_interactions['user_id'].unique()
movies = user_movie_interactions['item_id'].unique()

B.add_nodes_from(users, bipartite=0, type='user')
B.add_nodes_from(movies, bipartite=1, type='movie')
edges = list(user_movie_interactions.itertuples(index=False, name=None))
B.add_edges_from(edges)
```

```python
# Initialize the graph
G = nx.Graph()

# Add users to the graph
for _, row in users.iterrows():
    G.add_node(f'user_{row["user_id"]}', type='user', age=row['age'], gender=row['gender'], occupation=row['occupation'], zip_code=row['zip_code'])

# Add movies to the graph
for _, row in movies.iterrows():
    genres = row[6:].to_dict()
    genre_list = [k for k, v in genres.items() if v == 1]
    G.add_node(f'movie_{row["movie_id"]}', type='movie', title=row['title'], release_date=row['release_date'], genres=genre_list)

# Add genres to the graph
unique_genres = [col for col in movies.columns[6:]]
for genre in unique_genres:
    G.add_node(f'genre_{genre}', type='genre', name=genre)

# Add edges between movies and genres
for _, row in movies.iterrows():
    movie_node = f'movie_{row["movie_id"]}'
    genres = row[6:].to_dict()
    for genre, value in genres.items():
        if value == 1:
            G.add_edge(movie_node, f'genre_{genre}', type='belongs_to')

# Add edges between users and movies based on ratings
for _, row in ratings.iterrows():
    user_node = f'user_{row["user_id"]}'
    movie_node = f'movie_{row["item_id"]}'
    G.add_edge(user_node, movie_node, type='rated', rating=row['rating'], timestamp=row['timestamp'])
```

```python
# Draw the graph with node colors based on type
color_map = []
for node in G:
    if G.nodes[node]['type'] == 'user':
        color_map.append('blue')
    elif G.nodes[node]['type'] == 'movie':
        color_map.append('green')
    elif G.nodes[node]['type'] == 'genre':
        color_map.append('red')

plt.figure(figsize=(14, 8))
pos = nx.spring_layout(G, k=0.15)  # Adjust layout for better visualization
nx.draw(G, pos, node_color=color_map, with_labels=True, node_size=500, font_size=8)
plt.title('MovieLens Knowledge Graph')
plt.show()
```
